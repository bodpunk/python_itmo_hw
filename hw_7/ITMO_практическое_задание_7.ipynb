```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm
```

```python
base_url = 'https://lifehacker.ru/topics/technology' # базовая часть ссылки, позже в цикле будем добавлять к ней пагинацию
response = requests.get('https://lifehacker.ru/topics/technology/?page=1') # получаем контент первой страиниц
soup = BeautifulSoup(response.text, 'lxml') # инициализируем объект bs4 и задаем парсер lxml
```

```python
raw_items = soup.find_all('a', class_='lh-small-article-card__link')  # ищем все a-элементы с классом lh-small-article-card__link
links = [item.get('href') for item in raw_items] # получаем у ссылок только href-атрибут
links
```

```python
parsed_urls = []
```

```python
for page_num in range(1, 11):
    url = f'{base_url}/?page={page_num}' # подставляем нужный номер страницы для пагинации, дальше парсим каждую страницу аналогично коду выше
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'lxml')
    raw_items = soup.find_all('a', class_='lh-small-article-card__link')
    parsed_urls += [item.get('href') for item in raw_items]
```

```python
parsed_urls
```

```python
#возможно, это какая-то защита, но при парсинге ссылок на страницы, подсовываются ссылки такого вида:
#'https://burninghut.ru/onlajn-vishlisty/?utm_source=lifehacker.ru&utm_medium=referral&utm_campaign=teaser&utm_content=onlajn-vishlisty'
#(вероятно это реклама, я сам не вижу её на сайте)
#+ сюда же в цикл - удаляю заглушки, наткнулся на одну такую на сайте
#подытожу: грохаю всё, что начинается не со слеша, или содержит только заглушку-слеш
for x in range(len(parsed_urls) - 1, -1, -1):
  if parsed_urls[x][0] != '/' or parsed_urls[x] == '/':
    del parsed_urls[x]
```

```python
#стало чуть меньше 300 статей, но нам хватит)
len(parsed_urls)
```

```python
base_url = 'https://lifehacker.ru'
```

```python
response = requests.get('https://lifehacker.ru/chto-luchshe-elektricheskaya-ili-indukcionnaya-varochnaya-panel/')
soup = BeautifulSoup(response.text, 'lxml')
soup.find('article', class_='single-article__post-content').text
```

```python
#здесь на страницах по единственному h1, можно не использовать класс при поиске
soup.find('h1').text
```

```python
result = []

for url in tqdm(parsed_urls):
    article = {}

    article_url = f'{base_url}{url}' # подставляем часть ссылки, ведущую материал
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'lxml')

#в ссылки прокрадываются рекламные ссылки, имитирующие статьи (в т.ч. имитируют классы),
#но внутри у них нет элементов, из которых мы тащим текст, и в итоге получаем ошибку 'NoneType' object has no attribute 'text'
#поэтому сначала проверяем, что элементы есть, и только потом записываем из них контент
    if soup.find('h1') is not None and soup.find('article', class_='single-article__post-content') is not None:
      article['title'] = soup.find('h1').text # Получаем текст из h1 с классом article-card__title
      article['text'] = soup.find('article', class_='single-article__post-content').text # Получаем весь текст из article с классом single-article__post-content

      result.append(article)
```

```python
#проверяем сколько в итоге статей спарсили, должно быть чуть менее 300
len(result)
```

```python
#строим красивую табличку
pd.set_option('display.max_colwidth', 400)

data = pd.DataFrame(result)
data
```
