```python
# Задание специально или случайно неполное, речь про цену:
# по ТЗ нужна проверка на цену больше нуля,
# также по ТЗ ошибку возвращаем только в случае отрицательной цены;
# не описано требуемое поведение с ценой 0,
# для этого значения вывожу собственную ошибку

class Chocolate:
  def __init__(self, price: float, additive: str = ""):
    self.price = price
    self.additive = additive
    if self.price < 0:
      raise ValueError("Цена должна быть больше 0!")
    elif self.price == 0:
      raise ValueError("Бесплатных шоколадок не бывает!")

  def show_my_chocolate(self):
    if self.additive == "":
      print("Обычный шоколадный батончик")
    else:
      print(f"Шоколадный батончик, добавка: {self.additive}")

  def __str__(self):
    return f"Шоколадный батончик, {self.price} рублей"

sneakers = Chocolate(75, "Арахис")

nuts = Chocolate(20, "Нуга")

not_tasty_but_cheap = Chocolate(3.5, "Опилки")

milka = Chocolate(100, "")

bounty = Chocolate(73.29, "Кокос")
```
